El gradiente conjugado es conocido por ser es el método iterativo mas popular para resolver sistemas de equaciones lineales largos, es efectivo para sistemas de la forma <b>Ax=b</b>. 
Donde <b>x</b> es un vector desconocido, <b>b</b> is un vector conocido y <b>A</b> puede ser una matriz cuadratica positiva definida o indefinida.
<p>
<p>
En este caso se utilizarán las mayusculas para denotar matrices, las minúsculas para vectores y las letras  griegas para escalares. Podemos saber que <b>A</b> es una matriz <b>nxn</b> y <b>x</b> y <b>b</b> son vectores.</p>
$$  \begin{bmatrix}
 A_{11}  A_{12}  A_{13} \\ 
 A_{21}  A_{22}  A_{23} \\ 
 \vdots   \ddots   \vdots  \\ 
 A_{n1}  A_{n2}  A_{n3} 
\end{bmatrix}
\begin{bmatrix}
x_{1}\\ 
x_{2}\\ 
\vdots\\ 
x_{n}\\
\end{bmatrix}
=
\begin{bmatrix}
b_{1}\\ 
b_{2}\\ 
\vdots\\ 
b_{3}\\
\end{bmatrix} $$


<p>Para una mayor comprensión, se demostrará en python.</p>
<p>Primero se importan las librerías utilizadas:</p>

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import itertools
```

Para esta demostración se utilizará un ejemplo en 2d.

```python
A = np.matrix([[3.0, 2.0], [2.0, 6.0]])
b = np.matrix([[2.0], [-8.0]])  
c = 0.0
```

la forma cuadratica:
$$f(x)=\frac{1}{2}x^{T}Ax-b^{T}x+c$$
$$f'(x)= Ax-b$$
$$Ax-b =0$$

```python
def f(x, A, b, c):
    return float(0.5 * x.T * A * x - b.T * x + c)
```

Luego evaluamos la funcion en una cuadricula y la dibujamos con matplotlib:

```python
def bowl(A, b, c):
    fig = plt.figure(figsize=(10,8))
    qf = fig.gca(projection='3d')
    size = 20
    x1 = list(np.linspace(-6, 6, size))
    x2 = list(np.linspace(-6, 6, size))
    x1, x2 = np.meshgrid(x1, x2)
    zs = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            x = np.matrix([[x1[i,j]], [x2[i,j]]])
            zs[i,j] = f(x, A, b, c)
    qf.plot_surface(x1, x2, zs, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0)
    fig.show()
    return x1, x2, zs
```

```python
x1, x2, zs = bowl(A, b, c)
```

Luego si queremos ver el grafico de contorno de la misma forma cuadratica, aplicamos:

```python
def contoursteps(x1, x2, zs, steps=None):
    fig = plt.figure(figsize=(6,6))
    cp = plt.contour(x1, x2, zs, 10)
    plt.clabel(cp, inline=1, fontsize=10)
    if steps is not None:
        steps = np.matrix(steps)
        plt.plot(steps[:,0], steps[:,1], '-o')
    fig.show()
contoursteps(x1, x2, zs)
```
